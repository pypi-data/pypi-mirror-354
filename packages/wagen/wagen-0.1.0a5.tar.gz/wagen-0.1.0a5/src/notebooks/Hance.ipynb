## Adapt to error in prediction
NOW: **Imagine that we finished 'New Hance' with ACTUAL time of 330 minutes** (5.5 hours)

That would cause us to finish the day much later even if the rest of our predictions for the day remain accurate. But it could be that our first segment was slow because we are generally more tired than expected in which case the remainder of the day might "just as slow". 

I can think of several ways of recalculating the deviation of time to camp:
1. Keep the remainer of our pace predictions the same as they were
2. Multiply the paces for the remainder times a GIVEN factor. (GIVEN might be chosen from pre-calculated values)
3. Use the same average precentage change in pace predictions for remainder as we have seen up to this point
4. Use #3 times a fudge factor

```python
seg = 'New Hance' # we finished this Segment
act_here_mins = 330
pred_here_mins = seg_df.loc['New Hance','cs_mins']
pred_day_mins = seg_df.loc['New Hance','day_mins']
pred_rem_mins = seg_df.loc['New Hance','rem_mins']
pre = act_here_mins
pred_post = pred_day_mins - pred_here_mins
ua = tl.UpdateArrival(seg_df, seg)

display(eval(f'dict({ua.__repr__()})'))

print(f'Method 1: {ua.multiply_future(factor=1)} ')
print(f'Method 2: {ua.multiply_future(1.5)} ')
print(f'Method 3: {ua.future_as_previous(fudge=1)} ')
print(f'Method 4: {ua.future_as_previous(0.9)} ')
fap = ua.future_as_previous(1, act_here_mins)
print(f'Method 4: {fap} {fap/60:.1f} (hours) ')
print(f'{act_here_mins=} act_here_hrs={act_here_mins/60:.1f} '
      f'{pred_here_mins=:.0f} pred_here_hrs={pred_here_mins/60:.1f}'
      f'{act_here_mins/pred_here_mins:.0f} {pred_rem_mins=:.0f} {pre=} {pred_post=:.0f} ')
print()

# unchanged_future()
method = 'Method 1'
new_day_mins = pre + pred_post
print(f'{method}: Minutes to camp changed from {pred_day_mins=:.0f} to {new_day_mins=:.0f} ({pred_day_mins/60:.1f} to {new_day_mins/60:.1f}) ')

# multiply_future()
method = 'Method 2'
factor = 1.5
new_day_mins = pre + pred_post * factor
print(f'{method}: Minutes to camp changed from {pred_day_mins=:.0f} to {new_day_mins=:.0f} ({pred_day_mins/60:.1f} to {new_day_mins/60:.1f}) ')

# future_adjusted_like_previous()
method = 'Method 3'
factor = act_here_mins / pred_here_mins
new_day_mins = pre + pred_post * factor
print(f'{method}: Minutes to camp changed from {pred_day_mins=:.0f} to {new_day_mins=:.0f} ({pred_day_mins/60:.1f} to {new_day_mins/60:.1f}) ')

# future_like_previous_multiplied()
method = 'Method 4'
fudge = 0.9
factor = act_here_mins / pred_here_mins
new_day_mins = pre + pred_post * factor * fudge
print(f'{method}: Minutes to camp changed from {pred_day_mins=:.0f} to {new_day_mins=:.0f} new_day_hrs={new_day_mins/60:.1f} '
      f'({pred_day_mins/60:.1f} to {new_day_mins/60:.1f}) ')
```
