```python
import wagen.caltopo_io as cti
import wagen.timeline as tl
import pandas as pd
from pprint import pp
import datetime as dt
```

# Escalante Route
Down Tanner, up New Hance. Main obstacles are Papago wall (class 3) and Papago slide (i.e. "gully"

## Waypoints

```python
df, trip = cti.travelplan_multi_csv(csvfname='~/Downloads/travelplan.GC-Escalante.csv',
                                    caltopo_predictions=True)
print(trip)
with pd.option_context("display.max_columns", None):
    display(df.round(1))
```

```python
pd.DataFrame(trip)
```

```python
df.groupby('segment', sort=False).agg({'segment','last'})
```

## Timeline

```python
times_df, meta = tl.timeline(df, trip, 
                             start_time=dt.datetime.fromisoformat('2025-09-23'),
                             morning_daylight=1.5,
                             evening_daylight=4.0,
                             include_dt=False)
display(trip)
display(meta)
with pd.option_context("display.max_columns", None):
    display(times_df.round(1))
```

## Summary by day

```python
day_totals = df.groupby('day').agg({c:'sum' for c in ['distance','break_time','hours','aeg','ael']})
day_totals.round(1)
```

## Distance per segment (miles)

```python
pivot_df = df.pivot_table(index='day', columns=['segment'], aggfunc='sum', values='distance')
pivot_df.round(1)
```
